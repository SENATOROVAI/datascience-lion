1. Напишите алгоритм для расчета простых процентов от некоторой суммы.

 (P) – это сумма, от которой нужно рассчитать проценты.
 (r) – это процент, который нужно применить к исходной сумме.
 (t) – это период, за который нужно начислить простые проценты
  Расчет простых процентов (I) по формуле:
I = (P * r * t) / 100
  
1. Исходная сумма (P) = 1000 рублей
2. Годовая процентная ставка (r) = 10%
3. Время (t) = 3 года

Расчет простых процентов (I):
I = (1000 * 10 * 3) / 100 = 300 рублей


2. Напишите алгоритм для вычисления площади прямоугольника.

Алгоритм в виде шагов:
S = площадь прямоугольника
A = длина прямоугольника
B = ширина прямоугольника 
Формула для расчет — S = A * B


1. A = 5 единиц
2. B = 3 единицы
3. S = 5 × 3 = 15 единиц²
4. Результат: площадь прямоугольника равна 15 единиц².


3. Напишите алгоритм вычисления периметра круга. 

Алгоритм:
1. Определяем радиус (r) круга.
2. Используем формулу для расчета периметра круга:
   Периметр = 2 × π × r
   Где:
   π (пи) - математическая константа, равная примерно 3,14159.
3. Подставляем значение радиуса в формулу.
4. Вычисляем результат.
5. Представляем полученное значение как периметр круга.

Пример:
1. Радиус круга = 5 единиц
2. 2 × π ≈ 6,28318
3. Периметр = 6,28318 × 5 = 31,4159 единиц
4. Результат: периметр круга равен 31,4159 единиц.
