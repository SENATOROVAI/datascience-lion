Определить вероятность того, что серия наудачу выбранной облигации
не содержит одинаковых цифр, если номер серии может быть любым пятизначным
числом, начиная с 00001

 Будем решать по формуле P = n! / (n^n)
- P - вероятность, что номер серии не содержит одинаковых цифр
- n - количество цифр в номере серии ( 5)

P = 5! / (5^5)
P = 120 / 3125
P = 0.0384 = 3.84%

Какова вероятность угадать кодовый кодовый трехзначный номер, если он состоит: А) из четных цифр, причем цифры в числе не повторяются, б) из цифр больших 5, причем цифры в числе могут повторяться

А) Вероятность угадать трехзначный номер, состоящий из четных цифр, где цифры не повторяются:

1. Определим кол-во возможных четных цифр: 2, 4, 6, 8. (4 ШТ)
2. Надо понять общее кол-во вариантов. Мы знаем, что они не повторяются, а значит вариантов не так много (4 * 3 * 2 = 24). Типа, для первой цифры можно использовать 2,4,6,8. Для второй уже только 4,6,8. Для третьей (6,8)
3. Теперь можно найти вероятность по формуле кол-во благоприятных исходов (1) делим на общее кол-во возможных исходов (24)
P = 1/24 = 0.0417 или 4.17%

Б) Вероятность угадать трехзначный номер, состоящий из цифр больших 5, где цифры могут повторяться:

1.  Определим кол-во возможных цифр, больше 5: 6, 7, 8, 9. Всего 4 цифры.
2. Найдем общее кол-во возможных трехзначных номеров, состоящих из этих цифр (4 шт). Как и в примере А, только тут числа могут повторяться, поэтому в каждой ячейке может быть по 4 цифры. 
3. Ищем общее кол-во вариантов: 4 * 4 * 4 = 64
4. Теперь ищем вероятность по формуле P = 1/64 = 0.0156 или 1.56%
