# Введение в Python

#### Что такое Python?  

**Python** - это бесплатный интерпретируемый высокоуровневый язык программирования общего назначения с открытым исходным кодом.  

Был создан Гвидо ван Россумом  и впервые выпущен в 1991 г. Поддерживает несколько парадигм программирования (вкл. структурное и объекто-ориентированное)

- простота в изучении и использовании

> 2000 - Python 2.0 новые функции  
> 2008 - Python 3.0 сильно измененная версия, не имеющая полной обратной совместимости

в Python реализовано больше проверок ошибок, чем в C, в нем встроены многоуровневые типы данных: гибкие массивы и словари. Широкий диапазон задач. Python позволяет разделять программу на модели, которые можно повторно использовать в других программах. Большой встроенный набор стандартных модулей. Экономит время, так как нет компиляции. Интерпретатор можно использовать в интерактивном режиме. Позволяет писать программы компактно и читабельно:

#### Интерпретаторы Python  

**Интерпретатор** - разновидность компьютерной программы, которая непосредственно выполняет инструкции, написанные на каком-либо языке программирования или языке сценариев. Предварительная компиляция не требует.  

1. **IPython**: в комплекте идет редактор Jupyter Notebook. При установке Python через Anaconda автоматически устанавливается IPython.  

2. **CPython**: стандартная и наиболее широко используемая реализация языка программирования Python. Напи­санный на С и Python, интерпретатор CPython содер­жит много внешних функций С и связан с другими языками программирования.
CPython можно в какой-то степени считать компиля-
тором, поскольку перед интерпретацией он преобразует код Python в байт-код. В нем используется GIL (глобальная блокировка интерпретатора), которая ограни­ чивает возможности интерпретатор. Нужны расширения C для работы.  

3. **lronPython**  

Для платформы .NET. Есть поддержка динамической компиляции и встроенная интерактив­ ная консоль. IronPython напрямую интегрируется в среду разработки Visual Studio.  

4. **Jython**  

Работает на платформе Java. Обеспечивает поддержку статической и динамической компиляции. Позволяет импортировать  и использовать любые классы Java, словно это модуль Python.  

5. **РуРу**  

Реализован в RPython, который представляет собой ограниченное статически типизированное подмно­жество языка программирования Python.

Благодаря ЛТ-компилятору РуРу поддерживает сервер­ные части С, CLI и NM. Основная цель РуРу - обеспе-
чить максимальную совместимость с эталонной реализацией CPython и в то же время увеличить производительность. РуРу используется разработчиками Python, которые хотят увеличить производи­ тельность некоторых видов кода Python.  

6. **PythonNet**  

Позволяет выполнить почти бесшовную интеграцию Python с .NET CLR. Подход этого интерпретатора можно счи­тать противоположным тому, которому следует lronPython.
При использовании вместе с Mono, PythonNet позволяет встро­ енным в операционные системы установкам Python (кроме Windows) работать в среде .NET. PythonNet можно использо­вать совместно с IronPython без каких-либо проблем.  

7. Stackless Python  

Написан с использованием C  и Python. Наличие микропото­ ков. Эта возможность помогает избежать значительной части накладных: расходов, связанных с обычными потоками операционной системы  


#### 2.2.8 Объектная ориентированность  


Python поддерживает **процедурно-ориентированное программирование** и  **объекто-ориентированное программирование**. В процедурно-ориентированных языках программа состоит из процедур или функций, в которых заключены многократно используемые части программы.  В объектно-ориентированных языках про­грамма построена на объектах, которые включают в себе данные и функциональ­ность (очень мощный, но упрощенный стиль по сравнению с C++, Java)

> Функция - это блок кода, который выполняет определенную задачу или вычисление.  

#### 2.2.9. Расширяемость

### Jupyter Notebook  

#### Приложение Jupyter Notebook  

сконструирован как веб-приложение, подходящее для записи всего вычислительного процесса: разработки, документирования, выполнения кода, а также передачи результатов.  

**Компоненты**:  

- веб-приложение  
- документы Notebook  

#### Основные возможности веб-приложения  

- Работает с кодом в редакторе, встроенном в окно браузера, с автоматическим выделением синтаксиса цветом, отступами, заполнением, табуляцией и анализом.  
- Можно запустить код Python из боаузера и получить прикрепленные к коду результаты вычислений в форматах HTML, LaTeX, PNG, SVG  
- форматирование текста в браузере с использованием языка разметки Markdown   
- возможность простой интеграции математических формул в ячейки Markdown с помощью LaTeX и визуализации с помощью MathJax  


#### Документы Notebook  

Содержат входные и выходные данные интерактивного сеанса, дополнительный текст, который добавлен к коду. В файлах Notebook  содержится полная запись сеанса  работы в Веб-приложении, эти документы являются внутренними файлами JSON  и сохраняются с расширением ipynb. можно делиться файлами с другими пользователями.  

#### Структура документа Notebook   

 Последовательность ячеек. **Ячейка** - многострочное текстовое поле ввода. Выполнить содержимое: Shift + Enter / Play / Cell или Run. Поведение ячейки привыполнении зависит от ее типа.  
 Типы ячеек: code, ,markdown, ячейки без формата. По умолчанию ставится ячейка кода.  

 Написание кода с выделением синтаксиса и заполнением табуляцией. Язык программирования зависит от ядра, ядро по умолчанию Python запускает код Python.  

 Вывод: рисунки Matplotlib, таблицы HTML  

 **Ячейки Markdown** - разметка текста, указать, какой текст нужно выделить курсивом и т. д.  

 #### - заголовки  

 Может быть форматирование с помощью HTML- кода  

 Стандартная нотация LaTex - математика для внутристрочных выражений и больших формул, при выполнении ячейки автоматически конвертируется в HTML блягодаря MathJax.  

 **Неформатированные ячейки** не анализируются документом и при передаче nbconvent никак не преобразуются
